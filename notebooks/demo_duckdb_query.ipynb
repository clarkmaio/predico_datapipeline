# Load data
You can query data using SQL syntax using `duckdb` package.

`.parquet` files on Huggingface will work as a database.

```python
import duckdb
QUERY = f"SELECT * FROM 'hf://datasets/clarkmaio/elia_predico/target.parquet'"
conn = duckdb.connect()
result = conn.execute(QUERY).fetchdf()
result.head()
```
